```python
data["Cuisine"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Count")
plt.show()
```

```python
data["Average_Rating"].plot(kind="hist")

plt.title("Average Rating Distribution")
plt.xlabel("Average Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(data["Prices"], data["Average_Rating"])

plt.title("Price vs Average Rating")
plt.xlabel("Prices")
plt.ylabel("Average Rating")
plt.show()
```

```python
plt.boxplot(data["Prices"])

plt.title("Box Plot of Prices")
plt.ylabel("Prices")
plt.show()
```

```python
import seaborn as sns

corr = data[["Prices","Average_Rating","Votes","Total_Votes"]].corr()
sns.heatmap(corr, annot=True)

plt.title("Correlation Heatmap")
plt.show()
```

```python
pie = data["Is_Bestseller"].value_counts()

pie.plot(kind="pie", autopct="%1.1f%%")

plt.title("Best Seller")
plt.show()
```
